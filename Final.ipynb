```python
Car_price = dict(data.groupby('CarCompany').price.median())
```

```python
for i in Car_price.keys():
    
    if Car_price[i] <= 10000:
        Car_price[i]='low'
        
    elif (Car_price[i] > 10000) & (Car_price[i] < 20000):
        Car_price[i]='med'
        
    else:
        Car_price[i]='highend'
        
Car_price
```

```python
data.CarCompany = data.CarCompany.map(Car_price)
data.head()
```
